#### 1. Runnable Branch 

```python
# import Runnable Lambda and Runnable Branch 
from langchain_core.runnables import RunnableLambda,RunnableBranch

def is_math(input):
    return "math" in input.lower()

math_chain = RunnableLambda(lambda x: f"Solve math: {x}")
general_chain = RunnableLambda(lambda x: f"General response: {x}")


branch = RunnableBranch(
    (is_math, math_chain),
    general_chain
)
print(branch.invoke("What is 2 + 2?"))  # Output: Solve math: What is 2 + 2? False
print(branch.invoke("Do some math: What is 2 + 2?"))  # Output: Solve math: What is 2 + 2?
print(branch.invoke("What is the capital of France?"))  # Output: General response: What is the capital of France?
```

```python
from langchain_openai import AzureChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser 

import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

model = AzureChatOpenAI(
    azure_deployment=os.getenv("AZURE_OPENAI_DEPLOYMENT_NAME"),
    api_version=os.getenv("AZURE_OPENAI_API_VERSION"),
)

prompt = ChatPromptTemplate.from_template(
    "help me write a haiku about {topic}"
)
str = StrOutputParser()

chain = prompt | model | str
print(chain.invoke({"topic": "nature"}))
```
