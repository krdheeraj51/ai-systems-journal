```python
# prompt 
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert AI tutor."),
    ("user", "Explain {topic} in simple terms for a {audience}.")
])
```

```python
# LLM
from langchain_openai import AzureChatOpenAI
from dotenv import load_dotenv
import os

load_dotenv()
azure_endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
azure_api_key = os.getenv("AZURE_OPENAI_API_KEY")
azure_deployment_name = os.getenv("AZURE_OPENAI_DEPLOYMENT_NAME")
azure_api_version = os.getenv("AZURE_OPENAI_API_VERSION")


llm = AzureChatOpenAI(
    azure_deployment=azure_deployment_name,
    api_version=azure_api_version,
)
```

```python
from langchain_core.output_parsers  import StrOutputParser
str = StrOutputParser()
```

```python
chain = prompt | llm | str
response = chain.invoke({"topic": "Artificial Intelligence", "audience": "high school student"})
print(response)
```

```python
from langchain_core.prompts import FewShotChatMessagePromptTemplate
example = [
    {"input": "What is AI?", "output": "AI is machines simulating human intelligence."},
    {"input": "What is ML?", "output": "ML is a subset of AI that learns from data."}
]

example_prompt = ChatPromptTemplate.from_messages([
     ("user", "{input}"), 
     ("assistant", "{output}")]
    )
few_short_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=example)
final_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert AI tutor."),
    few_short_prompt,
    ("user", "{input}")
    ])

chain = final_prompt | llm | str
response = chain.invoke({"input": "What is deep learning?"})
print(response)
```
