```python
from langchain_openai import AzureChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.tools import tool
from langchain_core.output_parsers import StrOutputParser
import os 
from dotenv import load_dotenv
load_dotenv()


```

```python
## define a tool 

@tool 
def get_multiplytwoNumbers(num1: int, num2: int) -> int:
    """Multiplies two numbers together."""
    return num1 * num2      
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", """
You are an intelligent assistant.

Use tools when necessary.
Explain your reasoning briefly.
"""),
    ("user", "{input}")
])
azure_deployment = os.getenv("AZURE_OPENAI_DEPLOYMENT_NAME")
azure_api_version = os.getenv("AZURE_OPENAI_API_VERSION")
model = AzureChatOpenAI(
    azure_deployment=azure_deployment,
    api_version=azure_api_version,
)
model_with_tool = model.bind_tools([get_multiplytwoNumbers])
```

```python
chain = prompt | model_with_tool | StrOutputParser()
response = chain.invoke({"input": "What is 5 multiplied by 3?"})
print(response)
```
