# Objective 

Learn how PromptTemplate converts structured variables into a final prompt string 

### Key Concepts 
- input_variables
- template 
- format 

#### Steps

1. Create a template with 2 variables role, topic 
2. Format it with a values 
3. print the final prompt 

```python
from langchain_core.prompts import PromptTemplate
prompt = PromptTemplate(
    input_variables=["role","topic"],
    template="you are a {role} and you need to explain {topic} to me in simple terms.",
)
print(prompt.format(role="teacher", topic="quantum physics"))
```
